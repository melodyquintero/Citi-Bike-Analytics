```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

citibike_201701 = pd.read_csv("raw/JC-201701-citibike-tripdata.csv")
citibike_201702 = pd.read_csv("raw/JC-201702-citibike-tripdata.csv")
citibike_201703 = pd.read_csv("raw/JC-201703-citibike-tripdata.csv")
citibike_201704 = pd.read_csv("raw/JC-201704-citibike-tripdata.csv")
citibike_201705 = pd.read_csv("raw/JC-201705-citibike-tripdata.csv")
citibike_201706 = pd.read_csv("raw/JC-201706-citibike-tripdata.csv")
citibike_201707 = pd.read_csv("raw/JC-201707-citibike-tripdata.csv")

citibike_2017 = pd.concat([citibike_201701, citibike_201702,citibike_201703,citibike_201704,citibike_201705,citibike_201706,citibike_201707], ignore_index = True)


citibike_2017.tail()
citibike_2017.count()
```

```python

citibike_201801 = pd.read_csv("raw/JC-201801-citibike-tripdata.csv")
citibike_201802 = pd.read_csv("raw/JC-201802-citibike-tripdata.csv")
citibike_201803 = pd.read_csv("raw/JC-201803-citibike-tripdata.csv")
citibike_201804 = pd.read_csv("raw/JC-201804-citibike-tripdata.csv")
citibike_201805 = pd.read_csv("raw/JC-201805-citibike-tripdata.csv")
citibike_201806 = pd.read_csv("raw/JC-201806-citibike-tripdata.csv")
citibike_201807 = pd.read_csv("raw/JC-201807-citibike-tripdata.csv")

citibike_2018 = pd.concat([citibike_201801, citibike_201802,citibike_201803,citibike_201804,citibike_201805,citibike_201806,citibike_201807], ignore_index = True)

citibike_2018.head()

```

```python
citibike_2018.rename(columns={"tripduration": "Trip Duration", 
                                         "starttime": "Start Time", 
                                         "stoptime": "Stop Time",
                                         "start station id": "Start Station ID",
                                         "start station name": "Start Station Name",
                                         "start station latitude": "Start Station Latitude",
                                         "start station longitude": "Start Station Longitude",
                                         "end station id": "End Station ID",
                                         "end station name": "End Station Name",
                                         "end station latitude": "End Station Latitude",
                                         "end station longitude": "End Station Longitude",
                                         "bikeid": "Bike ID",
                                         "usertype": "User Type",
                                         "birth year": "Birth Year",
                                         "gender": "Gender"}, inplace=True)

citibike_2018.count()
```

```python
citibike_data=pd.concat([citibike_2017,citibike_2018], ignore_index = True)
citibike_data.to_csv("clean/citibike_data.csv")
```
